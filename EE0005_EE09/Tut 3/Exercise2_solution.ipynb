## Problem 2 : Statistical Summary

#### a) Find the Summary Statistics (Mean, Median, Quartiles etc) of SalePrice from the Numeric DataFrame.

```python
Saleprice = pd.DataFrame(train_data['SalePrice'])
Saleprice.describe()
```

#### b) Visualize the summary statistics and distribution of SalePrice using standard Box-Plot, Histogram, KDE.

```python
f, axes = plt.subplots(3, 1, figsize=(25, 40))
sb.boxplot(data = Saleprice, orient = "h", color = "violet", ax = axes[0])
sb.histplot(data = Saleprice, ax = axes[1])
sb.kdeplot(data = Saleprice, ax = axes[2])
```

#### c) Find the Summary Statistics (Mean, Median, Quartiles etc) of LotArea from the Numeric DataFrame.

```python
LotArea = pd.DataFrame(train_data['LotArea'])
LotArea.describe()
```

#### d) Visualize the summary statistics and distribution of LotArea using standard Box-Plot, Histogram, KDE.

```python
f, axes = plt.subplots(3,1, figsize=(34,30))
sb.boxplot(data = LotArea, orient = 'h', ax = axes[0])
sb.histplot(data = LotArea, ax = axes[1])
sb.kdeplot(data = LotArea, ax = axes[2])
```

#### e) Plot SalePrice (y-axis) vs LotArea (x-axis) using jointplot, and check the Correlation between the two.

```python
jointDF = pd.concat([LotArea, Saleprice], axis = 1).reindex(LotArea.index)
sb.jointplot(data = jointDF, x = "LotArea", y = "SalePrice", height = 12)
```

```python
jointDF.corr()
```

```python
sb.heatmap(jointDF.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```
